```python
import urllib.request
from bs4 import BeautifulSoup 
```

```python
url="https://www.flipkart.com/search?q=samsung%20galaxy%20s%20&marketplace=FLIPKART&otracker=start&as-show=on&as=off"
```

```python
page=urllib.request.urlopen(url)
```

```python
soup=BeautifulSoup(page,"html.parser")
```

```python
print(soup.prettify())
```

```python
containers=soup.findAll("div",attrs={"class":"_1-2Iqu row"})
```

```python
import csv
import pandas as pd
```

```python
for container in containers:
  model=container.find_all("div",attrs={"class":"_3wU53n"})
  clean_model=model[0].text
  ratings=container.find_all("div",attrs={"class":"hGSR34 _2beYZw"})
  clean_rating=ratings[0].text
  prices=container.find_all("div",attrs={"class":"_1vC4OE _2rQ-NK"})
  clean_price=prices[0].text
  #ROM=container.find_all("div",attrs={"class":"_3ULzGw"})
  #clean_ROM=ROM[0].text
  #new_rom=clean_ROM(" G ")
  print(clean_model)
  print(clean_rating)
  print(clean_price)
  
  final_rating=clean_rating.strip(" ★ ")
  price_split=clean_price.strip(" ₹ ")
  new_price=round(float(str(price_split).replace(",", "")),)
  print(clean_model)
  print(final_rating)
  print(new_price)
  with open('samsung.csv', 'a') as csv_file:
      writer = csv.writer(csv_file)
      writer.writerow([clean_model, final_rating,new_price ])
    
```

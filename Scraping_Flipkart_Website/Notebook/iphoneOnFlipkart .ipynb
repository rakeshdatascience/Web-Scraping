### We want to scrape model,rating & price

```python
for container in containers:
    product_name=container.find_all("div",attrs={"class":"_3wU53n"})
    product_clean = product_name[0].text
    ratings=container.find_all("div",attrs={"class":"hGSR34 _2beYZw"})
    ratings_clean = ratings[0].text
    price=container.find_all("div",attrs={"class":"_1vC4OE _2rQ-NK"})
    price_clean = price[0].text
    #n_ratings=container.find_all("span",attrs={"class":"_38sUEc"})
    #n_ratings_clean=reviews[0].text
    #final_n_ratings=n_ratings_clean.split(" R")[0]
    #stripChars = n_ratings_clean.translate(string.digits)  
    #final_n_ratings = n_ratings_clean.translate(stripChars)
    final_rating=ratings_clean.strip(" ★ ")
    price_split=price_clean.strip(" ₹ ")
    new_price=round(float(str(price_split).replace(",", "")),)
    
      
         
    
    print(new_price)
    print(product_clean)
    print(final_rating)
    
    with open('apple.csv', 'a') as csv_file:
      writer = csv.writer(csv_file)
      writer.writerow([product_clean, final_rating,new_price ])
    
      
    
 
```
